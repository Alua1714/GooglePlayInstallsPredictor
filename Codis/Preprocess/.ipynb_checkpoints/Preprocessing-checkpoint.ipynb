Read the csv

First of all, remove samples with less than 20 reviews, this should be done for test data also, since these observation are not relevant.

```python
dataMissingValues = dataOriginal.dropna()
print(dataMissingValues.shape)
data20reviews = dataMissingValues[dataMissingValues['Rating Count'] >= 20]
print(data20reviews.shape)
```

Remove columns that are not interesting to us in initially

```python
columnsToDrop = ['App Name', 'App Id', 'Minimum Android',
       'Minimum Installs', 'Developer Id', 'Developer Website','Developer Email', 
        'Privacy Policy', 'Scraped Time']
dataFilterColumns = data20reviews.drop(columns=columnsToDrop)
print(dataFilterColumns.shape)
```

Basic inspection of dataset

```python
dataFilterColumns.describe(include='all')
```

Let's search now for missing values in each colum

```python
key = ['Category', 'Rating', 'Installs', 'Free', 'Price', 'Currency', 'Maximum Installs', 'Size', 'Released', 'Last Updated', 'Content Rating', 'Ad Supported', 'In App Purchases', 'Editors Choice']

for column in key:
    print(column)
    print(dataFilterColumns[column].unique())
    print()
```

There are missing values in the column 'Currency' with value 'XXX'

```python
dataNoMissings1 = dataFilterColumns[dataFilterColumns['Currency'] != 'XXX']
print(dataNoMissings1['Currency'].unique())
```

There are some data in the 'Size' column with the value 'Varies with device', since this is not a numerical value, we have decided to erase them.

```python
dataNoMissings2 = dataNoMissings1[dataNoMissings1['Size'] != 'Varies with device']
print(sorted(dataNoMissings2['Size'].unique().astype(str)))
```

Data types conversion

```python
dataNoMissings2['Installs'] = dataNoMissings2['Installs'].str.replace('+', '')
dataNoMissings2['Installs'] = dataNoMissings2['Installs'].str.replace(',', '')
dataNoMissings2['Installs'].astype(float)
    
dataNoMissings2['Installs'].unique()
```

Now let's convert all prices to USD and erase the currency column (it will be redundant after this transformation)

```python
# Define exchange rates
exchange_rates = {'USD': 1.0, 'VND': 0.000043, 'GBP': 1.39, 'TRY': 0.12, 'EUR': 1.20}

# Function to convert price to USD
def convert_to_usd(row):
    currency = row['Currency']
    price = row['Price']
    exchange_rate = exchange_rates[currency]
    return round(price * exchange_rate, 2)

# Apply the function to create a new column 'Price_USD'
dataNoMissings2['Price'] = dataNoMissings2.apply(convert_to_usd, axis=1)
```

```python
dataTransformed1 = dataNoMissings2.drop(columns=['Currency'])
```

```python
dataTransformed1['Size'] = dataTransformed1['Size'].str.replace('G', '000000000')
dataTransformed1['Size'] = dataTransformed1['Size'].str.replace('M', '000000')
dataTransformed1['Size'] = dataTransformed1['Size'].str.replace('k', '000')
dataTransformed1['Size'] = dataTransformed1['Size'].str.replace('.', '')
dataTransformed1['Size'] = dataTransformed1['Size'].str.replace(',', '')
dataTransformed1['Size'].astype(float)
    
dataTransformed1['Size'].unique()
```

Now, the number of days from the released date and last update will be computed

```python
dataTransformed2 = dataTransformed1
# Convert 'Released' column to datetime with the correct format
dataTransformed2['Released'] = pd.to_datetime(dataTransformed1['Released'], format='%b %d, %Y')

# Define the target date 'Jun 15, 2021'
target_date = pd.to_datetime('Jun 15, 2021')

# Calculate the number of days elapsed
dataTransformed2['Days_Elapsed'] = (target_date - dataTransformed2['Released']).dt.days

# Update 'Released' column with the number of days elapsed
dataTransformed2['Released'] = dataTransformed2['Days_Elapsed']

# Optionally, you can drop the 'Days_Elapsed' column if you don't need it anymore
dataTransformed2 = dataTransformed2.drop(columns=['Days_Elapsed'])

sorted(dataTransformed2['Released'].unique())
```

```python
dataTransformed3 = dataTransformed2

# Convert 'Released' column to datetime with the correct format
dataTransformed3['Last Updated'] = pd.to_datetime(dataTransformed2['Last Updated'], format='%b %d, %Y')

# Define the target date 'Jun 15, 2021'
target_date = pd.to_datetime('Jun 15, 2021')

# Calculate the number of days elapsed
dataTransformed3['Days_Elapsed'] = (target_date - dataTransformed3['Last Updated']).dt.days

# Update 'Released' column with the number of days elapsed
dataTransformed3['Last Updated'] = dataTransformed3['Days_Elapsed']

# Optionally, you can drop the 'Days_Elapsed' column if you don't need it anymore
dataTransformed3 = dataTransformed3.drop(columns=['Days_Elapsed'])

sorted(dataTransformed3['Last Updated'].unique())
```

Set all types to the desired ones

```python
key = ['Rating', 'Installs', 'Price', 'Size', 'Released', 'Last Updated', 'Maximum Installs']

for i in key:
    dataTransformed3[i] = dataTransformed3[i].astype(float)
    
dataTransformed3.dtypes
```

Add the new new calculated variables

```python
X = dataTransformed3
Y = X['Installs']
X = X.drop(columns='Installs')

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=0)


X_train['Installs'] = y_train
dataSynthetic = X_train
dataSynthetic['Download'] = dataSynthetic['Installs']/dataSynthetic['Released']
```

```python
NormalizedData = dataLinearized6
floatColumns = NormalizedData.select_dtypes(include=['float']).columns
floatColumns
for column in floatColumns:
    # Calculate mean and variance
    mean = np.mean(NormalizedData[column])
    variance = np.var(NormalizedData[column])
    NormalizedData[column] = (NormalizedData[column] - mean) / np.sqrt(variance)
```
